```python
# Grouping and Aggregation
# Group the filtered DataFrame by ProductCategory
grouped_df = df_filtered.groupby('ProductCategory')

# Compute the total Quantity sold and total TotalSale for each category
grouped_df_sum = grouped_df_sum_sorted
# Sort the grouped result in descending order based on the total TotalSale
grouped_df_sum_sorted = grouped_df_sum.sort_values(by='TotalSale', ascending=False)
```

```python
# Identify and print the top 3 product categories with the highest total sales
print("\nTop 3 product categories with the highest total sales:")
print(grouped_df_sum_sorted.head(3))
```

```python
# Multi-Level Grouping
# Group the filtered data by both StoreID and ProductCategory
grouped_df_multi = df_filtered.groupby(['StoreID', 'ProductCategory'])
```

```python
# Calculate the average TotalSale per transaction for each group
average_totalsale = grouped_df_multi['TotalSale'].mean()
```

```python
# Store-Level Insights
# For each store, determine which product category has the highest average TotalSale
store_insights = average_totalsale.groupby('StoreID').idxmax()
```

```python
# Prepare and print a summary DataFrame

summary_df = pd.DataFrame({'StoreID': store_insights.index,
                             'TopProductCategory': store_insights.values,
                             'Average TotalSale': average_totalsale[store_insights.values]})
```

```python
print("\nStore-level insights:")
print(summary_df)
```
