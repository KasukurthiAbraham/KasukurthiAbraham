```python
def AnalyzeFrance():
  matchesPlayed = df['home_team'], value_counts()
  print(f'Total Number of matches played by France is :{matchesPlayed["France"]}')
```

```python
AnalyzeFrance()
```

```python
def AnalyzeTown() :
  Town = df['home_team'].value_counts()
  for _ in Town.index:
    print(f'Total Number of matches played by {_} is :{Town[_]}')
```

```python
AnalyzeTown()
```
